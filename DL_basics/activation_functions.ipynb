## Activation Functions in Neural Networks

Activation functions are used in neural networks to introduce **non-linearity**, allowing the network to learn complex patterns. Common activation functions include:

1. **Step Function**
   - Outputs 0 if input < 0, else 1
   - Simple but not differentiable → rarely used in practice

2. **Sigmoid**
   - Formula: σ(x) = 1 / (1 + exp(-x))
   - Output: (0, 1)
   - Pros: smooth, good for probabilities
   - Cons: vanishing gradient for large/small x, slow convergence

3. **Tanh**
   - Formula: tanh(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))
   - Output: (-1, 1)
   - Pros: zero-centered
   - Cons: still suffers from vanishing gradients

4. **ReLU (Rectified Linear Unit)**
   - Formula: f(x) = max(0, x)
   - Pros: fast, simple, reduces vanishing gradient
   - Cons: "dying ReLU" problem (neurons stuck at 0)

5. **Leaky ReLU**
   - Formula: f(x) = x if x > 0 else α*x (α small, e.g., 0.01)
   - Pros: solves dying ReLU problem
   - Cons: introduces small negative slope (may affect learning slightly)

### Key Points to Remember
- Activation functions allow **non-linear transformations**.
- Choice of activation affects **learning speed** and **performance**.
- ReLU and variants are the most popular in deep learning today.
- Sigmoid/tanh are mostly used in **output layers** for probabilities or certain architectures (like LSTM gates).


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)
```

```python
def step (x):
    return np.where(x >= 0, 1, 0)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,3,1)
plt.plot(x, step(x), label='Step', color='blue')
plt.title('Step Function')
plt.grid()
```

```python
def sigmoid (x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.subplot(2,3,2)
plt.plot(x, sigmoid(x), label='Sigmoid', color='orange') 
plt.title('Sigmoid Function')
plt.grid()
```

```python
def tanh (x):
    return np.tanh(x)
```

```python
plt.subplot(2,3,3)
plt.plot(x, tanh(x), label='Tanh', color='green')
plt.title('Tanh Function')
plt.grid()
```

```python
def leaky_relu (x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)
```

```python
plt.subplot(2,3,4)
plt.plot(x, leaky_relu(x), label='Leaky ReLU', color='red')
plt.title('Leaky ReLU Function')    
plt.grid()
```

```python
def relu (x):
    return np.maximum(0, x)
```

```python
plt.subplot(2,3,5) 
plt.plot(x, relu(x), label='ReLU', color='purple')
plt.title('ReLU Function')
plt.grid()
```

```python
# all in one 
plt.figure(figsize=(12, 10))

# Step
plt.subplot(3, 2, 1)
plt.plot(x, step(x), color='blue')
plt.title("Step Function")
plt.grid(True)

# Sigmoid
plt.subplot(3, 2, 2)
plt.plot(x, sigmoid(x) , color='orange')
plt.title("Sigmoid Function")
plt.grid(True)

# Tanh
plt.subplot(3, 2, 3)
plt.plot(x, tanh(x), color='green')
plt.title("Tanh Function")
plt.grid(True)

# ReLU
plt.subplot(3, 2, 4)
plt.plot(x, relu(x), color='purple')
plt.title("ReLU Function")
plt.grid(True)

# Leaky ReLU
plt.subplot(3, 2, 5)
plt.plot(x, leaky_relu(x), color='red')
plt.title("Leaky ReLU Function (alpha=0.01)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
